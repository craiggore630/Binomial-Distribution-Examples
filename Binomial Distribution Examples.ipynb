```python
import plotly.express as px
r_vals = range(11)
dist = [binom.pmf(r, 10, 0.5) for r in r_vals ] 

fig = px.histogram(x = r_vals, y = dist, nbins = 11, width = 500, text_auto = True)
fig.update_traces(marker_line_width=1,marker_line_color="black")
fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 1
    )
)
fig.show()
```
